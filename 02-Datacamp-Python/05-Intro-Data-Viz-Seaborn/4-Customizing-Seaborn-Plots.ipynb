# 4. Customizing Seaborn Plots

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## Figure style

- includes background and axes
- preset options: `white`, `dark`, `whitegrid`, `darkgrid`, `ticks`
- ```sns.set_style()```

````python
# "How often do you listen to your parents' advice?". Now let's change the style and palette to make this plot easier to interpret.

survey_data = pd.read_csv("datasets/young-people-survey-responses.csv")
survey_data = survey_data.rename(columns={"Parents' advice": "Parents Advice"})

# Set the style to "whitegrid" to help the audience determine the number of responses in each category.
sns.set_style("whitegrid")

# Create a count plot of survey responses
# category_order = ["Never", "Rarely", "Sometimes", 
#                   "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count")
            # order=category_order)

plt.show()
````

## Palette

- changes the color of the main elements
- ```sns.set_palette()```
- use preset palettes or create custom ones
- Preset: 
  - diverging pallettes: "RdBu", "PRGn", "RdBu_r", "PRGn_r"
  - sequential palettes (for continuing scale): "Greys", "Blues", "PuRd", "GnBu"

````python
purple_palette = sns.color_palette("Purples")
sns.palplot(purple_palette)
````

````python
rdbu_palette = sns.color_palette("RdBu", 15)
sns.palplot(rdbu_palette)
````

````python
# Reverse

rdbu_reverse_palette = sns.color_palette("RdBu_r", 15)
sns.palplot(rdbu_reverse_palette)
````

````python
# Set the color palette to the sequential palette named "Purples"
sns.set_style("whitegrid")
sns.set_palette("Purples")

# Create a count plot of survey responses
# category_order = ["Never", "Rarely", "Sometimes", "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count")
            # order=category_order)

plt.show()
````

````python
# Change the color palette to the diverging palette named "RdBu"

sns.set_style("whitegrid")
sns.set_palette("RdBu")

# Create a count plot of survey responses
# category_order = ["Never", "Rarely", "Sometimes", "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count")
            # order=category_order)

plt.show()
````

## Changing the scale

- figure "context" changes the scale of the plot elements and labels
- ```sns.set_context()```
- smallest to largest: "paper", "notebook", "talk", "poster"
- the default is "paper"

````python
# Does the percentage of people reporting that they feel lonely vary depending on how many siblings they have? Let's find out using a bar plot, while also exploring Seaborn's four different plot scales ("contexts").

# Set the scale ("context") to "paper", which is the smallest of the scale options.
sns.set_context("paper")

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, 
            kind="bar")

plt.show()
````

````python
# Change the context to "notebook" to increase the scale.

sns.set_context("notebook")

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, 
            kind="bar")

plt.show()
````

````python
# Change the context to "talk" to increase the scale.

sns.set_context("talk")

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, 
            kind="bar")

plt.show()
````

````python
# Change the context to "poster", which is the largest scale available.

sns.set_context("poster")

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, 
            kind="bar")

plt.show()
````

## Titles and Labels

- FaceGrid and AxesSubplot

| Object Type | Plot Types | Characteristics |
|-------------|------------|-----------------|
| `FacetGrid`  | `relplot()`, `catplot()` | can create subplots |
| `AxesSubplot` | `scatterplot()`, `countplot()`, ... | only creates a single plot |



````python
# Seaborn plot functions create two different types of objects: FacetGrid objects and AxesSubplot objects. The method for adding a title to your plot will differ depending on the type of object it is.

# In the code provided, we've used relplot() with the miles per gallon dataset to create a scatter plot showing the relationship between a car's weight and its horsepower. This scatter plot is assigned to the variable name g. Let's identify which type of object it is.

mpg = pd.read_csv("datasets/mpg.csv")

# Identify what type of object plot g is and assign it to the variable type_of_g.

# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g)

# Print type
print(type_of_g)
````

````python
# Add the following title to this plot: "Car Weight vs. Horsepower".

# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
````
